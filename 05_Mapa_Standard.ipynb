# Mapa Standard

```
q = 2 #@param {type:"integer"}
K = 2 #@param {type:"number"}
max_iter = 10000 #@param {type:"integer"}
```

```
import math as m
import matplotlib.pyplot as plt
```

```
x0, p0 = 0.001, 0.001 # Condiciones iniciales

dot_x = []
dot_y = []

for i in range(max_iter):
  
    p = p0 + (K/(2*m.pi*q)) *m.sin(2*m.pi*x0)
    x = x0 + q*p
     
    x0, p0 = x, p

    dot_x.append(x%1)
    dot_y.append(p%1)
```

```
plt.xlabel("x")
plt.ylabel("P")
plt.title("Mapa standard reducido")

plt.plot(dot_x, dot_y, '.', markersize=3, color='red')
plt.show()
```
