```
import numpy as np
import scipy.integrate as integrate
import scipy.special as special
import matplotlib.pyplot as plt
```

```
def pend(y,t):

	q,p = y 			              # vector Y
	dydt = [p,(-1)*np.sin(q)] 	# vector Y'

	return dydt
```

```
y0 = [np.pi,0]      		    	# vector condición inicial.
t = np.arange(0, 100, 0.1)
```

```
sol = integrate.odeint(pend, y0, t)
```

```
plt.xlabel("$\phi$")
plt.ylabel("p")

q0 = np.arange(np.pi/2, np.pi, 0.1)

for i in q0:
	y0 = [i,0]
	sol = integrate.odeint(pend, y0, t)
	plt.plot(sol[:,0], sol[:,1], color='blue')

plt.show()
```

```
h = np.arange(0, 5, 0.01)
f1 = lambda h: np.pi/(2*special.ellipk(np.sqrt((h+1)/2)))
f2 = lambda h: np.pi*np.sqrt((h+1)/2)/(2*special.ellipk(np.sqrt(2/(h+1))))

w = np.piecewise(h, [h<1, h>1], [f1, f2])

plt.xlabel("h")
plt.ylabel("$\omega$(h)")

plt.plot(h,w)
plt.show()
```

```
h = np.arange(0.9,1.01,0.0011)
w = np.piecewise(h, [h<1, h>1], [f1, f2])

plt.xlabel("h")
plt.ylabel("$\omega$(h)")

plt.plot(h,w)
plt.show()
```
